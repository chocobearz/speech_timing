```python
import pandas as pd
import re
import os
import ast
```

```python
scriptDirectory = r"C:\\Users\\ptut0\\Documents\\speech_timing\\TEDLIUM_release-3\\speaker-adaptation\\train\\txt\\clean\\"
writeDirectory = "C:\\Users\\ptut0\\Documents\\speech_timing\\TEDLIUM_release-3\\speaker-adaptation\\train\\"
data = pd.read_csv("C:\\Users\\ptut0\\Documents\\speech_timing\\TEDLIUM_release-3\\speaker-adaptation\\train\\word_length.csv")
#chunks = pd.read_csv("C:\\Users\\ptut0\\Documents\\speech_timing\\TEDLIUM_release-3\\speaker-adaptation\\train\\chunk_length.csv")
```

```python
for file in os.listdir(scriptDirectory):
  if ".txt" in file:
    with open(scriptDirectory+file, 'r') as f:
      script = f.read()
      if script[0] == ',':
        script = script[2:]
      script = script.replace("\n","")
      script = script.replace("’","'")
      script = script.replace(",.",'.')
      script = script.replace("  "," ")
      script = script.replace(" . ",". ")
      script = script.replace(" “",", ")
      script = script.replace("” ",", ")
      script = script.replace("”","")
      script = script.replace("-"," ")
      text_file = open(scriptDirectory+"clean\\"+file, "w")
      text_file.write(script)
      text_file.close()
      f.close()
```

```python
def cleanScript(script):
    sentences = re.split('(?<=[.!?]) +', script)
    
    # Group sentences into chunks of up to 400 words
    chunked_sentences = []
    current_chunk = ''
    for sentence in sentences:
        # If adding the current sentence will exceed 400 words, append the current chunk to the list of chunked sentences
        if len(current_chunk.split()) + len(sentence.split()) > 400:
            chunked_sentences.append(current_chunk.strip())
            current_chunk = ''
        # Add the current sentence to the current chunk
        current_chunk += sentence + " "
    # Append the final chunk to the list of chunked sentences
    if current_chunk:
        chunked_sentences.append(current_chunk.strip())
    
    return chunked_sentences
```

```python
k = 0
for index, row in data.iterrows():
  print(row['filename'])
  pauseLengths = ast.literal_eval(row['pause_lengths'])
  wordLengths = ast.literal_eval(row['word_lengths'])
  script_location = scriptDirectory+str(row['filename'])+".txt"
  with open(script_location, "r") as f:
    script = f.read()
  scriptList = cleanScript(script)
  m = 0
  print(index)  
  if len(wordLengths) == len(script.split()):
    print(k)
    k +=1
    if len(wordLengths)-1 == len(pauseLengths):
      for chunk in scriptList:
        dataframe = pd.DataFrame({'filename' : [], 'word_lengths' : [], 'pause_lengths' : [], 'script' : []})
        binLengths = {'filename': row['filename'], 'word_lengths' : [], 'pause_lengths' : [], 'script' : ''}
        words = chunk.split()
        n = len(words)
        binLengths['word_lengths'] = (wordLengths[m:n+m])
        binLengths['pause_lengths'] = (pauseLengths[m:n+m])
        binLengths['script'] = chunk
        m += n
        dataframe = pd.concat([dataframe, pd.Series(binLengths).to_frame().T], ignore_index=True)
        if(len(dataframe['word_lengths'][0]) != len(dataframe['script'][0].split())):
           print("oops")
        dataframe.to_csv(writeDirectory+"chunk_length.csv", mode='a', header=not os.path.exists(writeDirectory+"chunk_length.csv"))
```
